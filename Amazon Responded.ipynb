```python
dff = df.select("tweet_created_at","user_screen_name","user_id_str")
```

```python
dff.show()
```

```python
import pyspark
import pyspark.sql.functions as fuc

# Simplify the column 'tweet_created_at' as "Month Day Year"
dfs = dff.withColumn('tweet_created_at', fuc.concat(df.tweet_created_at.substr(5, 6), df.tweet_created_at.substr(26, 5)))
```

```python
dfs.show(5)
```

```python
#step 1
df1 = dfs.groupBy("user_screen_name", "user_id_str").agg(fuc.countDistinct("tweet_created_at"))
df1.show()
```

```python
#
daily_post_count = dfs.groupBy(['user_screen_name', 'user_id_str', 'tweet_created_at']).agg(countDistinct('user_id_str').alias('num_posts'))
daily_post_count.show()
```

```python
#
active_days_count = daily_post_count.groupBy(['user_screen_name', 'user_id_str']).agg(countDistinct('tweet_created_at').alias('num_active_days'))
active_days_count.show()
```

```python
#
daily_active_users = active_days_count.filter('num_active_days >= 5').select('user_screen_name', 'user_id_str')
daily_active_users.show()
```

```python
#
num_active_users = daily_active_users.count()
print("number of active user", num_active_users)
```

```python
# join the first file with experiment file
df5= daily_active_users.join(df4,daily_active_users.user_id_str == df4.user_id,"rightouter")
```

```python
#rename the coloumn
df5= df5.select("user_screen_name","user_id")
df5= df5.withColumnRenamed("user_id","user_id_str")
df5.show()
```

```python
from pyspark.sql.functions import when
df5=df5.withColumn("whether_active",when(df5.user_screen_name.isNotNull(),"yes").otherwise("no"))
```

```python
df5.show()
```

```python

experiment_user =df5.select("user_id_str","whether_active")
experiment_user.show()

experiment_user.write.csv('experiment_user.csv')
```

```python
#calculate the percentage of active user and print out the result
total_users = experiment_user.count()
active_users = experiment_user.filter("whether_active == 'yes'").count()
active_percentage = active_users / total_users * 100

# print the result
print("Out of the {} experiment targets, {} are active users, which is {:.2f}% of the total.".format(total_users, active_users, active_percentage))
```
